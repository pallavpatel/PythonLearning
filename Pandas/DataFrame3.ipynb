# A review of .apply() Method on single column

```python
bond = pd.read_csv("c:/Pallav/pandas/jamesbond.csv" , index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
def convert_num_to_string_and_add_Million(number):
    return str(number) + "Millions"

bond["Box Office"] = bond["Box Office"].apply(convert_num_to_string_and_add_Million)
```

```python
bond["Budget"] = bond["Budget"].apply(convert_num_to_string_and_add_Million)
```

```python
bond.head(3)
```

# The .apply() Method with rows

```python
bond = pd.read_csv("c:/Pallav/pandas/jamesbond.csv" , index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
def good_movies(row):
    actor = row[1]
    budget = row[4]
    if actor == "Pierce Brosnan":
        return "The Best"
    if actor == "Roger Moore" and budget > 40:
        return " Enjoyable"
    else:
        return " No Clue"

    
bond.apply(good_movies, axis = "columns")
```
