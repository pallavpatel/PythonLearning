# The .astype() Method

```python
nba = pd.read_csv("c:/pallav/pandas/nba.csv").dropna(how = "all")
nba.tail()
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace =  True)
nba.head()
```

```python
nba.dtypes

# or 
nba.info()
```

```python
# Salary is float data type. Converting float to int data type

nba["Salary"] = nba["Salary"].astype("int") 
nba.head()

nba.info()

# You can reasign back to same column from int to float.
```

```python
#nunique gives unique values from a column.

nba["Position"].nunique()
```

```python
# category is unique in pandas. It is not present in python. It will take unique values from column and assign category.
# It will decrease memory usage. You can check that by executing inf0 command and see memory usage.
# There will no change in column value.

nba["Position"] = nba["Position"].astype("category")
```

```python
nba.info()
```

```python
nba["Team"].nunique()
nba["Team"] = nba["Team"].astype("category")
```

# Sort a DataFrame with .sort_values() Method Part 1

```python
nba = pd.read_csv("c:/pallav/pandas/nba.csv")
nba.head()
```

```python
# This will sort entire DataFrame as per column which we are sorting. 
# By default all NaN (Null) Values will go last irrespective of ascending order

nba.sort_values("Name" , ascending = False)

nba.sort_values("Age" , ascending = True)


nba.sort_values("Salary", na_position = "first").head()  # Keeping all Null values on first place irrespective or ascending order.
```

# Sort a DataFrame with .sort_values() Method Part 2

```python
nba = pd.read_csv("c:/pallav/pandas/nba.csv")
nba.head()
```

```python
nba.sort_values(["Team","Name"]) # This will sort first on Team and then Name in alphabetical order.

nba.sort_values(["Team","Name"], ascending = False).head() # rever order for both column

nba.sort_values(["Team","Name"], ascending = [True, False]) # Here Team will be ascending order but Name will be descending order.

# You can use inplace parameter to make changes in original DataFrame.
```

# rank values with .rank() Method

```python
nba = pd.read_csv("c:/pallav/pandas/nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary"].rank() # Here it will rank salary based on lowest first
nba["Salary"].rank(ascending = False) # highest will get lowest rank
nba["Salary"].rank(ascending = False).astype("int") # Converting to int type of rank
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")  # assigning it to new column

nba.sort_values("Salary", ascending = False).head()

# Same values will have same ranking.

# if same rank is assigned to then next number in rank will be skipped.
```
