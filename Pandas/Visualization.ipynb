# Bar Chart

```python
google = data.DataReader(name = "GOOG" , data_source = "google" , start = "2004-01-01" , end = "2016-12-31")
google.head()
```

```python
def rank_performance(stock_price):
    if stock_price < 200:
        return "Poor"
    elif stock_price >=200 and stock_price <500:
        return "Satisfactory"
    else:
        return "Steller"
```

```python
google["Close"].apply(rank_performance).value_counts()
```

```python
google["Close"].apply(rank_performance).value_counts().plot()
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "line")
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "Bar")
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "Barh")
```

# Pie Chart

```python
apple = data.DataReader(name = "AAPL" , data_source = "google" , start = "2012-07-01" , end = "2016-12-31")
apple.head()
```

```python
apple["Close"].mean()
```

```python
def apple_rank_performance(stock_price):
    if stock_price > 95:
        return "Above Average"
    else:
        return "Below Average"
```

```python
apple["Close"].apply(apple_rank_performance).value_counts().plot(kind = "pie" , legend = True)
```
