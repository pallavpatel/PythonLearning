# Categorical Plot

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x ='sex' , y= 'total_bill', data = tips) # gives average total bills for male and female
```

```python
sns.barplot(x ='sex' , y= 'total_bill', data = tips, estimator = sum)
```

```python
import numpy as np

sns.barplot(x ='sex' , y= 'total_bill', data = tips, estimator = np.std)
```

```python
sns.countplot(x= 'sex' , data = tips)  # it will provide count
```

```python
sns.boxplot(x = 'day' , y= 'total_bill', data = tips)  # box plot
```

```python
sns.boxplot(x = 'day' , y= 'total_bill', data = tips , hue = 'smoker') 
```

```python
sns.violinplot(x = 'day' , y = 'total_bill' , data = tips, hue = 'sex' , split = True)
```

```python
sns.stripplot(x= 'day' , y ='total_bill' , data = tips )
```

```python
sns.stripplot(x= 'day' , y ='total_bill' , data = tips , hue = 'sex')
```

```python
sns.stripplot(x= 'day' , y ='total_bill' , data = tips , hue = 'sex', jitter= True)
```

```python
sns.swarmplot(x = 'day' , y = 'total_bill' , data = tips)  # combination of violine and striplot
```

# Matrix Plot

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc , annot= True)
```

```python
sns.heatmap(tc , annot= True , cmap = 'coolwarm')
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index = 'year' , columns= 'month' , values= 'passengers')
```

```python
sns.heatmap(fp, cmap = 'coolwarm', linecolor = 'white' , linewidths= 3)
```

# Regression Plot

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = tips) # lm is linear model
```

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = tips, hue = 'sex') # lm is linear model
sns.plt.show()
```

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = tips , col = 'sex' , row = 'time')

sns.plt.show()
```

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = tips , col = 'day' ,  hue = 'time')

sns.plt.show()
```

```python
sns.lmplot(x = 'total_bill' , y = 'tip' , data = tips , col = 'day' ,  hue = 'time', aspect = 0.6 , size = 10)

sns.plt.show()
```

# Grids

```python
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
sns.plt.show()
```

```python
g = sns.PairGrid(iris)
g.map(sns.plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.plt.scatter)
g.map_lower(sns.kdeplot)
sns.plt.show()
```

# Seaborn Exercises

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x = 'fare' , y = 'age', data = titanic, size=10)
```

```python
sns.distplot(titanic['fare'] , kde= False, color = 'red', bins = 30)
```

```python
sns.boxplot(x='class' , y= 'age', data = titanic, palette='rainbow')
```

```python
sns.swarmplot(x = 'class' , y = 'age' , data = titanic, palette='Set2')
```

```python
sns.countplot(x = 'sex' , data = titanic )
```

```python
sns.heatmap(titanic.corr(), cmap = 'coolwarm')
sns.plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data = titanic, col = 'sex')

g.map(sns.distplot, 'age', kde = False)
```

```python
g = sns.FacetGrid(data = titanic, col = 'sex')

g.map(plt.hist , 'age' )
```
