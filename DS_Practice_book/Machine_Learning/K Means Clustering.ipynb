```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(
                n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101
                    )
```

```python
data[0].shape
```

```python
data
```

```python
plt.scatter(data[0][:,0] , data[0][:,1])
```

```python
plt.scatter(data[0][:,0] , data[0][:,1] , c = data[1])
```

```python
plt.scatter(data[0][:,0] , data[0][:,1] , c = data[1], cmap = 'rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_  # in real life this label will not be available. This is unsupervised learning. 
```

```python
fig , (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize = (10,4))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0] , data[0][:,1] , c= kmeans.labels_ , cmap = 'rainbow' )

ax2.set_title('Original')
ax2.scatter(data[0][:,0] , data[0][:,1] , c= data[1] , cmap = 'rainbow' )
```

```python

kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])
kmeans.cluster_centers_
kmeans.labels_

fig , (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize = (10,4))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0] , data[0][:,1] , c= kmeans.labels_ , cmap = 'rainbow' )

ax2.set_title('Original')
ax2.scatter(data[0][:,0] , data[0][:,1] , c= data[1] , cmap = 'rainbow' )
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans  =  KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private' , axis = 1))
```

```python
kmeans.verbose
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
def convert_private(private):
    if private == 'Yes':
        return 0
    else:
        return 1
```

```python
df['Cluster'] = df['Private'].apply(convert_private)
```

```python
df.tail()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'] , kmeans.labels_))
```

```python
print(classification_report(df['Cluster'] , kmeans.labels_))
```
