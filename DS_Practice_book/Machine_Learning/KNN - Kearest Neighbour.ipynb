```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("C:/Pallav/Python-Data-Science-and-Machine-Learning-Bootcamp/Python-Data-Science-and-Machine-Learning-Bootcamp/Machine_Learning_Sections/K-Nearest-Neighbors/Classified Data",
                  index_col = 0
                  )
```

```python
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df1.drop('TARGET CLASS', axis = 1))
```

```python
scaled_features = scaler.transform(df1.drop('TARGET CLASS', axis = 1))
```

```python
scaled_features
```

```python
df_fit = pd.DataFrame(scaled_features , columns=df1.columns[:-1])
```

```python
df_fit.head()
```

```python
df1.head()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X = df_fit  # or you can pass scaled_features
y = df1['TARGET CLASS']
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test, pred))
```

```python
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,4))
plt.plot(range(1,40) , error_rate , color = 'Blue' , linestyle = 'dashed' , marker = 'o' , markerfacecolor = 'red',
        markersize = 10)
plt.title("Error rate Vs K value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

```python
knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train, y_train)
pred = knn.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test, pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=34)
knn.fit(X_train, y_train)
pred = knn.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test, pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("C:/Pallav/Python-Data-Science-and-Machine-Learning-Bootcamp/Python-Data-Science-and-Machine-Learning-Bootcamp/Machine_Learning_Sections/K-Nearest-Neighbors/KNN_Project_Data"
                                 )
```

```python
df1.head()
```

```python
sns.pairplot(df1, hue='TARGET CLASS' ,palette='RdBu')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
proj_scaler = StandardScaler()
```

```python

proj_scaler.fit(df1.drop('TARGET CLASS', axis = 1))
```

```python
proj_scaled_features = proj_scaler.transform(df1.drop('TARGET CLASS', axis = 1))
```

```python
proj_scaled_features
```

```python
proj_df1_fit = pd.DataFrame(proj_scaled_features , columns=df1.columns[:-1])
```

```python
proj_df1_fit.head()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X = proj_df1_fit
y = df1['TARGET CLASS']

X_train, x_test, y_train, y_test = train_test_split(X, y , test_size = 0.3, random_state = 101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test,pred))
```

```python
error_rate = []

for i in range(1,200):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python

plt.figure(figsize=(15,4))
plt.plot(range(1,75) , error_rate[:74] , color = 'Blue' , linestyle = 'dashed' , marker = 'o' , markerfacecolor = 'red',
        markersize = 10)
plt.title("Error rate Vs K value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

```python
knn = KNeighborsClassifier(n_neighbors=32)
knn.fit(X_train, y_train)
pred = knn.predict(x_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test,pred))
```
