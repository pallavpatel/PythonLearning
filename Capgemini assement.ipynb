```python


Year = []
Ceremony = []
Award = []
Winner = []
Film = []
Name = []

data = open("C:/Pallav/Python_practice/input_1.txt","r") 

replace("and" , ',')

for _, row in data.iterrows():
    Year = row.Year
    Ceremony = row.Ceremony
    Award = row.Award
    Winner = row.Winner
    Film = row.Film
    
    for actor_name in data.Name:
        Year.append(Year)
        Ceremony.append(Ceremony)
        Award.append(Award)
        Winner.append(Winner)
        Film.append(Film)
        Name.append(actor_name)
        
new_data = pd.DataFrame({
    "Year": Year,
    "Ceremony": Ceremony,
    "Award": Award,
    "Winner" : Winner,
    "Film" : Film,
    "Name" : Name
})


new_data
```
